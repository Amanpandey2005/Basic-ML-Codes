```python
import requests
import pandas as pd
import time

all_pages = []

for i in range(1, 525):
    print(f"Fetching page {i}...")   # ⭐ Progress print
    
    # API request with error-handling
    url = f"https://api.themoviedb.org/3/movie/top_rated?api_key=657926941dacbaae8d7437e49941eee8&language=en-US&page={i}"
    
    try:
        response = requests.get(url)
        response.raise_for_status()   # error check
        data = response.json().get("results", [])
        
        if data:   # avoid empty pages
            temp_df = pd.DataFrame(data)[[
                'id', 'title', 'overview','popularity','release_date','vote_average','vote_count'
            ]]
            all_pages.append(temp_df)

    except Exception as e:
        print(f"Error on page {i}: {e}")

    time.sleep(0.20)   # ⭐ prevents rate-limit freeze

# Combine all pages
df = pd.concat(all_pages, ignore_index=True)
print("Done! Final shape:", df.shape)
```
